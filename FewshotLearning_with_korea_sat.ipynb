```python
total_score = 0
```

```python
for n_paragraph in range(len(kor_sat_df['paragraph'])):
  for n_problems in range(len(kor_sat_df['problems'][n_paragraph])):
    text = kor_sat_df['paragraph'][n_paragraph]
    question = kor_sat_df['problems'][n_paragraph][n_problems]['question']
    options = kor_sat_df['problems'][n_paragraph][n_problems]['choices']
    best_answer = find_best_answer(text, question, options)
    if str(best_answer) == str(kor_sat_df['problems'][n_paragraph][n_problems]['answer']):
      total_score += kor_sat_df['problems'][n_paragraph][n_problems]['score']
    print(f"예측 정답 : {best_answer}, 실제 정답 : {kor_sat_df['problems'][n_paragraph][n_problems]['answer']}, 총 점수 : {total_score}")
```
